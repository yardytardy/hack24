```python
import csv
import numpy as np
import pandas as pd
from tabulate import tabulate
import matplotlib.pyplot as plt
consoldata = pd.read_csv('consoldata.csv')

varnames = (consoldata.iloc[:,1])

with_s =[]
while not with_s:
    muni = input("Please Enter Your Municipality: ").lower()
    with_s = [x for x in varnames if x.startswith(muni)]
    if not with_s:
        print("\nThat Municipality Does Not Exist. Try Again: ")
if len(with_s) == 1:
    muni = with_s[0]
    print("\nGreat! Your Municipality is " + muni.capitalize() + ".")
else:
    print("\nThere are Multiple with that Name: ")
    for x in range(len(with_s)):
        print(str(x + 1) + " " + with_s[x])
    muninum = int(input("Pick the Municipality You are Trying to Select: 1 through " + str(len(with_s)) + ": "))
    if muninum > 0 and muninum < len(with_s) + 1:
        muni = with_s[muninum - 1]
        print("Great! Your Municipality is " + muni.capitalize() + ".")
    else:
        while muninum < 1 or muninum > len(with_s):
            muninum = int(input("Insert a Valid Number" + ": "))
        muni = with_s[muninum - 1]
        print("Great! Your Municipality is " + muni.capitalize() + ".")
        
        
rownum = consoldata.index[consoldata.iloc[:,1] == muni].tolist()[0]
print("\nThe County your Municipality is Located in is " + consoldata.iloc[rownum, 2] + ".")
print("Your Municipality's Electric Utility is " + consoldata.iloc[rownum, 57] + ".")
print("Your Municipality's Gas Utility is " + consoldata.iloc[rownum, 63] + ".")
print("Your Municipality's Metropolitan Planning Organization is " + consoldata.iloc[rownum, 80] + ".")

types = ["Population Characteristics", "Electricity by Sector", "Natural Gas by Sector", "Primary Heating Fuels", 
         "GHG Emissions", "Means of Transportation for Commuting", "VMT by Vehicle Type", "EV Ownership",
        "Solar Installations", "Residential Housing Data", "Units by Structure Type", "Residential EE Participation",
        "Commercial EE Participation"]
print("\nSelect the Data you Want: \n")
for x in range(13):
    print(str(x + 1) + " " + types[x])
outnum = int(input("\nPick the Data You are Trying to See, From 1 through 13. If You Would Like to Stop, Please Type the Number 20. "))

while outnum != 20:
    if outnum == 1:
        table = [["% White", consoldata.iloc[rownum, 6] * 100], 
                ["% Black", consoldata.iloc[rownum, 7] * 100], 
                ["% Asian, Pacific Islander, and Hawaiian", consoldata.iloc[rownum, 8] * 100],
                ["% Other", consoldata.iloc[rownum, 9] * 100], 
                ["% Hispanic of Latino Origin", consoldata.iloc[rownum, 10] * 100],
                ["Households", consoldata.iloc[rownum, 11]], 
                ["Median Household Income", consoldata.iloc[rownum, 12]],
                ["Percent of Population in Poverty", consoldata.iloc[rownum, 13] * 100], 
                ["NJ DCA MRI Score", consoldata.iloc[rownum, 15]],
                ["NJ DCA MRI Rank", consoldata.iloc[rownum, 16]], 
                ["Square Miles", consoldata.iloc[rownum, 4]]]
        print("\nCommunity Profile Data in " + muni.capitalize() + ":\n")
        print(tabulate(table))
    elif outnum == 2:
        names = ["Residential Electricity", "Commercial Electricity", "Industrial Electricity", "Street Lighting Electricity"]
        values = [float(consoldata.iloc[rownum, 58])/1000, float(consoldata.iloc[rownum, 59])/1000,
                  float(consoldata.iloc[rownum, 60])/1000, float(consoldata.iloc[rownum, 61])/1000]
        plt.figure(figsize=(9, 3))
        plt.bar(names, values)
        plt.suptitle("Electricity Purchased by Sector for " + muni.capitalize() + ".")
        plt.ylabel("Thousands of kWh")
        plt.show()
    elif outnum == 3:
        names = ["Residential Gas", "Commercial Gas", "Industrial Gas", "Street Lighting Gas"]
        values = [float(consoldata.iloc[rownum, 64])/1000, float(consoldata.iloc[rownum, 65])/1000,
                  float(consoldata.iloc[rownum, 66])/1000, float(consoldata.iloc[rownum, 67])/1000]
        plt.figure(figsize=(9, 3))
        plt.bar(names, values)
        plt.suptitle("Natural Gas Purchased by Sector for " + muni.capitalize() + ".")
        plt.ylabel("Thousands of Therms")
        plt.show()
    elif outnum == 4:
        names = ["Utility Gas", "Electricity", "Bottled, Tank, or LP Gas", "Fuel Oil, Kerosene, etc.", "Wood", "Other"]
        values = [float(consoldata.iloc[rownum, 42]), float(consoldata.iloc[rownum, 43]),
                 float(consoldata.iloc[rownum, 44]), float(consoldata.iloc[rownum, 45]),
                 float(consoldata.iloc[rownum, 46]), float(consoldata.iloc[rownum, 47])]
        plt.suptitle("Residential Heating Fuels for " + muni.capitalize() + ".")
        plt.pie(values, labels = names)
        plt.show() 
    elif outnum == 5:
        names = ["Residential Electricity", "Commercial Electricity", "Industrial Electricity", "Street Lighting Electricity",
                "Residential Gas", "Commercial Gas", "Industrial Gas", "Street Lighting Gas", "Other Heating Fuels", "Road Vehicles"]
        values = [float(consoldata.iloc[rownum, 69]), float(consoldata.iloc[rownum, 70]),
                 float(consoldata.iloc[rownum, 71]), float(consoldata.iloc[rownum, 72]),
                 float(consoldata.iloc[rownum, 73]), float(consoldata.iloc[rownum, 74]),
                 float(consoldata.iloc[rownum, 75]), float(consoldata.iloc[rownum, 76]),
                 float(consoldata.iloc[rownum, 77]), float(consoldata.iloc[rownum, 78])]
        plt.suptitle("Greenhouse Gas Emissions for " + muni.capitalize() + " In Metric Tons of CO2.")
        plt.pie(values, labels = names)
        plt.show() 
    elif outnum == 6:
        names = ["Car, Truck, Van", "Public Transit", "Taxicab", "Motorcycle", "Bicycle", "Walked", "Other", "Worked at Home"]
        values = [consoldata.iloc[rownum, 49], consoldata.iloc[rownum, 50],
                 consoldata.iloc[rownum, 51], consoldata.iloc[rownum, 52],
                 consoldata.iloc[rownum, 53], consoldata.iloc[rownum, 54],
                 consoldata.iloc[rownum, 55], consoldata.iloc[rownum, 56]]
        plt.suptitle("Means of Commuting to Work for " + muni.capitalize() + ".")
        plt.pie(values, labels = names)
        plt.show() 
    elif outnum == 7:
        names = ["Buses", "Combination Trucks", "Light Trucks", "Motorcycles", "Passenger Cars", "Single Unit Trucks"]
        values = [(float(consoldata.iloc[rownum, 83])/1000) + (float(consoldata.iloc[rownum, 90])/1000) + (float(consoldata.iloc[rownum, 93])/1000),
                 (float(consoldata.iloc[rownum, 81])/1000) + (float(consoldata.iloc[rownum, 82])/1000),
                 (float(consoldata.iloc[rownum, 84])/1000) + (float(consoldata.iloc[rownum, 88])/1000),
                 (float(consoldata.iloc[rownum, 85])/1000) + (float(consoldata.iloc[rownum, 86])/1000),
                 float(consoldata.iloc[rownum, 87])/1000,
                 (float(consoldata.iloc[rownum, 91])/1000) + (float(consoldata.iloc[rownum, 92])/1000)]
        plt.figure(figsize=(10, 3))
        plt.bar(names, values)
        plt.suptitle("Vehicle Miles Traveled by Vehicle for " + muni.capitalize() + ".")
        plt.ylabel("Thousands of Miles")
        plt.show()
    elif outnum == 8:
        table = [["Estimated Total Vehicles", consoldata.iloc[rownum, 95]],
                ["# of EVs", consoldata.iloc[rownum, 96]],
                ["% Electric Vehicles", consoldata.iloc[rownum, 97] * 100]]
        print("\nElectric Vehicle Data in " + muni.capitalize() + ":\n")
        print(tabulate(table))
    elif outnum == 9:
        print("\nThe number of existing solar installations in " + muni.capitalize() + " is " + str(consoldata.iloc[rownum, 101]) + ".")
    elif outnum == 10:
        table = [["Median Household Income", consoldata.iloc[rownum, 12]],
                ["Percent of Population in Poverty", consoldata.iloc[rownum, 13] * 100],
                ["% Owner Occupied", consoldata.iloc[rownum, 21] * 100],
                ["% Renter Occupied", consoldata.iloc[rownum, 22] * 100]]
        print("\nResidential Household Data in " + muni.capitalize() + ":\n")
        print(tabulate(table))
    elif outnum == 11:
        names = ["1-Unit Detached", "1-Unit Attached", "Duplex", "3-4 Units", "5-9 Units", "10-19 Units", "20 or More Units", "Mobile Homes"]
        values = [float(consoldata.iloc[rownum, 33]),
                 float(consoldata.iloc[rownum, 34]),
                 float(consoldata.iloc[rownum, 35]),
                 float(consoldata.iloc[rownum, 36]),
                 float(consoldata.iloc[rownum, 37]),
                 float(consoldata.iloc[rownum, 38]),
                 float(consoldata.iloc[rownum, 39]),
                 float(consoldata.iloc[rownum, 40])]
        plt.figure(figsize=(12, 3))
        plt.bar(names, values)
        plt.suptitle("Housing Structures by Units for " + muni.capitalize() + ".")
        plt.ylabel("Units")
        plt.show()
    elif outnum == 12:
        table = [["Units in Structures with 4 Units or Less", consoldata.iloc[rownum, 103]],
                ["Total Completed Projects", consoldata.iloc[rownum, 104]],
                ["Current Lifetime %", float(consoldata.iloc[rownum, 105]) * 100]]
        print("\nResidential EE Program Participation Data in " + muni.capitalize() + ":\n")
        print(tabulate(table))
    elif outnum == 13:
        table = [["# of C&I Parcels", consoldata.iloc[rownum, 106]],
                ["Total Completed Projects", consoldata.iloc[rownum, 107]],
                ["Current Lifetime %", float(consoldata.iloc[rownum, 108]) * 100]]
        print("\nCommercial EE Program Participation Data in " + muni.capitalize() + ":\n")
        print(tabulate(table))
    elif outnum < 1 or outnum > 13:
        print("OOPS! Try Again! ")
    outnum = int(input("\nPick the Data You are Trying to See, From 1 through 13. If You Would Like to Stop, Please Type the Number 20. "))
```
